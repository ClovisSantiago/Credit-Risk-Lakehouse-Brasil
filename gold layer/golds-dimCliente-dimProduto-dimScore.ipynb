```python
%sql
-- ============================================================================
-- Tabela......: gold.dim_cliente
-- Camada......: Gold
-- Objetivo....: Dimensão de clientes
-- Fonte.......: silver.dim_cliente
-- Grão........: 1 registro por cliente
-- Responsável.: Clovis Santiago
-- ============================================================================
CREATE OR REPLACE TABLE dev_credit_risk.gold.dim_cliente
AS

SELECT

    id_cliente,
    nome_cliente,
    idade,
    sexo,
    estado_civil,
    faixa_renda,
    uf

FROM dev_credit_risk.silver.dim_cliente;
```

```python
%sql
-- ============================================================================
-- Tabela......: gold.dim_produto_credito
-- Camada......: Gold
-- Objetivo....: Dimensão de produtos de crédito
-- Fonte.......: silver.dim_produto_credito
-- Grão........: 1 registro por produto de crédito
-- Responsável.: Clovis Santiago
-- ============================================================================
CREATE OR REPLACE TABLE dev_credit_risk.gold.dim_produto_credito
AS

SELECT

    id_produto,
    modalidade_credito,
    taxa_juros,
    prazo_maximo_meses

FROM dev_credit_risk.silver.dim_produto_credito;
```

```python
%sql
-- ============================================================================
-- Tabela......: gold.dim_score
-- Camada......: Gold
-- Objetivo....: Dimensão de score
-- Fonte.......: silver.dim_score
-- Grão........: 1 registro por score
-- Responsável.: Clovis Santiago
-- ============================================================================
CREATE OR REPLACE TABLE dev_credit_risk.gold.dim_score
AS

SELECT

    id_score,
    faixa_score,
    risco

FROM dev_credit_risk.silver.dim_score;
```
