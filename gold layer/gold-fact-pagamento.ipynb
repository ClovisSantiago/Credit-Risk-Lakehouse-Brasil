```python
%sql
CREATE OR REPLACE TABLE dev_credit_risk.gold.fact_pagamento_credito AS

SELECT
    fp.id_pagamento,
    fp.id_contrato,

    fc.id_cliente,
    fc.id_produto,
    fc.id_score,

    fp.dt_vencimento,
    fp.dt_pagamento,
    fp.dt_competencia,
    fp.valor_pago,
    fp.dias_atraso,

    CASE
        WHEN fp.dias_atraso > 0 THEN 1
        ELSE 0
    END AS flag_pagamento_atrasado,

    CASE WHEN fp.dias_atraso >= 90 THEN 1 
        ELSE 0 
    END AS flag_pagamento_default_90,

    CASE
        WHEN fp.dias_atraso >= 90 THEN 'Default 90+'
        WHEN fp.dias_atraso > 0 THEN 'Pago em atraso'
        WHEN fp.dias_atraso <= 0 THEN 'Pago em dia'
        ELSE 'Indefinido'
    END AS status_pagamento

FROM dev_credit_risk.silver.fact_pagamento fp

LEFT JOIN dev_credit_risk.gold.fact_carteira_credito fc
    ON fp.id_contrato = fc.id_contrato

WHERE fp.dias_atraso <> 999;
```
