```python
%sql
CREATE OR REPLACE TABLE dev_credit_risk.gold.fact_carteira_credito

USING DELTA

AS

SELECT

    id_contrato,

    id_cliente,

    id_produto,

    id_score,

    dt_contratacao,

    date(date_trunc('MONTH', dt_contratacao)) AS safra_contratacao,

    valor_contratado,

    prazo_meses,

    taxa_juros_contratada as taxa_juros

FROM dev_credit_risk.silver.fact_contrato;
```

```python
%sql
SELECT

MIN(dt_contratacao),

MAX(dt_contratacao)

FROM dev_credit_risk.gold.fact_carteira_credito;
```
