### Dimensão de cliente dim_cliente
-----------------------------------

#### Consultando os dados do arquivo de clientes

```python
df_cliente =(
  spark.read.option("header", "True")
  .option("inferSchema", "True")
  .csv("/Volumes/dev_credit_risk/bronze/data_source/dim_cliente_raw.csv")
)
df_cliente.display()
```

####Criando tabela e escrevendo dados na tabela delta dim_cliente

```python
path = "dev_credit_risk.bronze."
df_cliente.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(path + "dim_cliente")
```

####Validando nova tabela delta dim_cliente

```python
%sql
select * from dev_credit_risk.bronze.dim_cliente
```
